```python
import numpy as np
import matplotlib.pyplot as plt
import cPickle
from ae import *
%matplotlib inline
```

```python
with open ("ae.pkl", "rb") as fin:
    ae = cPickle.load(fin)
images = load_MNIST_images('train-images-idx3-ubyte')
```

```python
img = images[:, 1].reshape((28, 28))
plt.subplot(121)
plt.title("original image")
plt.imshow(img, cmap="gray")
img_reconstruct = ae.reconstruct(img.reshape((-1, 1)))
plt.subplot(122)
plt.title("reconstructed image of CPU Autoencoder")
plt.imshow(img_reconstruct.reshape((28, 28)), cmap="gray")
plt.show()
```

```python
plt.subplot(121)
plt.title("original image")
plt.imshow(img, cmap="gray")
plt.subplot(122)
plt.title("reconstructed image of GPU Autoencoder")
img_reconstruct = np.genfromtxt("result.txt", delimiter=" ")
plt.imshow(img_reconstruct, cmap="gray")
plt.show()
```
